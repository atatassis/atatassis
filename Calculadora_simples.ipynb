```python
# Vamos fazer uma calculadora básica:

# Primeiro criei uma função chamada "Calculadora", que é onde nosso código vai ser desenvolvido.
# Depois, organizei as váriaveis. Em "Calculo" irei perguntar que tipo de operação a pessoa vai querer usar.
# Em "numero1" e "numero2" ordenará a sequência dos dígitos na operação. O usuário deverá informá-los.
def calculadora():
  calculo = input("Qual o tipo de operação você quer usar? ")
  numero1 = int(input())
  numero2 = int(input())

# Aqui, fará a soma dos números informados se o usuário digitar "Adição"
# Leitura: Se cálculo for igual a Adição, retorne numero1 +(mais) numero2
  if calculo == "Adição":
    return numero1 + numero2

# Aqui, fará a subtração dos números informados se o usuário digitar "Subtração"
# Leitura: Se cálculo for igual a Subtração, retorne numero1 -(menos) numero2
  elif calculo == "Subtração":
    return numero1 - numero2

# Aqui, fará a multiplicação dos números informados se o usuário digitar "Multiplicação"
# Leitura: Se cálculo for igual a Multiplicação, retorne numero1 *(vezes) numero2
  elif calculo == "Multiplicação":
    return numero1 * numero2

# Aqui, fará a divisão dos números informados se o usuário digitar "Divisão"
# Leitura: Se cálculo for igual a Divisão, retorne numero1 //(dividido) numero2
  elif calculo == "Divisão":
    return numero1 // numero2

# Para finalizar, chamamos a função dentro do print, para que seja executada e imprima os cálculos.
print(calculadora())

```
