```python
# Criaremos funções que retornem as operaçãos de "a" e "b":
def soma(a, b):
    return a + b

def subtracao(a, b):
    return a - b

def multiplicacao(a, b):
    return a * b

def divisao(a, b):
    if b != 0:
        return a // b
    else:
        return "Não é possível dividir por 0" # Como não é possível uma divisão por 0, retornará a mensagem e o programa reinicia.

# Enquanto o bloco for verdadeiro, as operações serão feitas.
# Caso o usuário inserir qualquer digito que fuja da verdade, o programa resultará em erro: "ValueError".
while True:
    num1 = int(input())
    if num1 == 0:
        print(None)
        break
    operacao = input()
    num2 = int(input())

# Chamamos a função dentro do print, imprimindo de acordo com a operação que o usuário escolher.

    if operacao == '+':
        print(soma(num1, num2))
    elif operacao == '-':
        print(subtracao(num1, num2))
    elif operacao == '*':
        print(multiplicacao(num1, num2))
    elif operacao == '/':
        print(divisao(num1, num2))
    else:
        print("Operação inválida! Tente novamente.")
```
